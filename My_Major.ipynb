# **전공에 관해**  

&nbsp;&nbsp;내 전공은 조금 특이한 경우로, 어쩌면 기하급수적인 학생수 감소를 겪고 있는 현재에 이르러선 이런 일이 도리어 평범한 것인지도 모르겠는데, [학과 홈페이지](http://korean2.hannam.ac.kr/sub1/menu_1.html)의 표현을 따라 적자면 '국어국문·창작학과'로, 기존의 '국어국문학과'와 '문예창작학과'가 통합된 것이다. '국어국문·창작학과'는 명칭 뿐 아니라 두 학과의 교육과정도 계승하여, 각각 '국어국문트랙'과 '문예창작트랙'으로 분류하고 있다.  
&nbsp;&nbsp;'국어국문학과'는 1969년 대전대학('한남대학교'의 옛 이름)에 신설되었으며, '문예창작학과'는 1999년 한남대학교 야간 대학에 개설돼 2005년에 주간으로 전환됐다. '문예창작학과'는 국어국문학부에 속한 별개의 학과로 취급 및 운영되었으나, 2015년에 이르러 두 학과는 '국어국문·창작학과'라는 이름으로 통합되었다.

&nbsp;&nbsp;내가 '국어국문·창작학과'에 적을 둔 것은 2018년인데, 두 학과가 하나로 합쳐져 운영된지 겨우 삼년째였고, 졸업을 앞둔 4학년들을 위한 후속적인 지원과 같은 까닭으로 종전의 제도를 유지하는 탓에, '국어국문·창작학과'로서의 제도적인 정비는 충분히 이뤄지지 않은 시기였다. 학과의 학생회장이 둘이었고 신입생 정원은 80명이었다.  
&nbsp;&nbsp;학과의 통합을 지시했던 총장께서 2018학년도를 마지막으로 퇴임한 이후 새로 부임한 총장은 학교의 재정문제를 이유로 문과대학 소속 학과를 개편하고자 했다. 이 과정에서 '철학상담학과'가 이름과 소속을 바꾸었고, '일어일문학과'와 '불어불문학과'가 통합됐다. '국어국문·창작학과'는 아예 존폐의 기로에 서게 됐는데, 이에 몇 가지 조건을 수용하는 것으로 현재에 이르렀다 들은 바가 있다.

&nbsp;&nbsp;그 조건 중에는 교육과정의 개편이 있었고 2020년을 기준으로 강의의 명칭과 교육과정이나 교재 따위가 전면적으로 개편·개정되었는데, 마침 이 시기에 코로나의 유행은 물론 나는 군 복무를 시작해, 복학 이후 학과의 분위기가 아주 뒤바뀌어서 내 1·2학년 동안의 체험으론 현재의 '국어국문·창작학과'를 충분히 설명할 수 있을성싶지 않다.  
&nbsp;&nbsp;내가 당시에 보고 들은 것은 오로지 순문학이었으나, 최근 '국어국문·창작학과'에선 순문학의 방법론은 거의 가르치지 않고 웹소설 같은 '장르문학'이란 걸 다룬다는데, 내 개인적인 의견이자 결론이지만, 둘은 '문학'이란 조어만을 공유할 뿐 공통분모가 없는 별개의 것이라 보아도 좋을 듯하다.
